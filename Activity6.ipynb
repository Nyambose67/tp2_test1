# DATA VISUALIZATION
plotting a histogram graph using jersy number column: mean=19.534636
std = 14.273515 total_row=571

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=10, scale=5.6, size=571)
plt.hist(data,bins=10, color='blue', alpha=0.7)
plt.title("HISTOGRAM OF GOAL DISTRIBUTED TO EACH JERSY NUMBER")
plt.ylabel("Goals per Match")
plt.xlabel("Jersy number")
plt.show()
```

Creating a Scatter plot for Appearances vs Losses

```python
df = pd.read_csv('C:\\Users\\user\\Desktop\\Test2 Prep\\dataset - 2020-09-24.csv')

Losses = [n for n in df["Losses"].head(10)]
appearances= [n for n in df["Appearances"].head(10)]

plt.scatter(appearances,Losses)
plt.title("Scatter Plot of Matches lost vs Appearances")
plt.xlabel("Appearances")
plt.ylabel("Matches lost")
plt.show()
```

Creating a Bar Chart comparing Appearences for the last five players 

```python
players = [n for n in df["Name"].tail()]
appearances = [n for n in df["Appearances"].tail()]

plt.figure(figsize=(12, 5))
plt.bar(players, appearances, color='black', alpha=0.7)
plt.title('Game Appearances of each first five Players')
plt.xlabel('Players')
plt.ylabel('Appearances')
plt.show()
```

```python
Names = [n for n in df["Name"].head(10)]
goals= [n for n in df["Goals"].head(10)]

plt.figure(figsize=(7,8))
plt.pie(goals, labels=Names, autopct='%1.1f%%', startangle=15)
plt.title("First 10 Players's Goals Scored")
plt.show()
```

Creating a Box Plot 



```python
Ycards = [n for n in df["Yellow cards"]]
plt.figure(figsize=(6, 8))
plt.boxplot(Ycards, vert=True, patch_artist=True, meanline=True, showmeans=True)
plt.title("Distribution of Yellow Cards")
plt.ylabel("Yellow Cards")
plt.grid(True)
plt.show()
```
